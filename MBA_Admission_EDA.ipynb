### **Adjusting Missing Values**

```python
df["race"]=df['race'].fillna(df['race'].mode()[0])
```

```python
df.isnull().sum()/len(df)*100
```

```python
df["admission"]=df['admission'].fillna(df['admission'].mode()[0])
```
