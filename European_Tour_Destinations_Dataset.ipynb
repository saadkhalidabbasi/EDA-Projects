## **Visualzing our Data and Exploring through Graphs**

```python
fig=px.pie(df,names="Country",title="Country Distribution")
fig.show()
```

```python
fig=px.pie(df,names="Cost of Living",title="Cost of Living Distribution")
fig.update_traces(textposition="inside",textinfo="percent+label")
fig.show()
```

```python
sns.countplot(data=df,x='Cost of Living')
```

```python
# Group by the columns
df.groupby('Cost of Living').size()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Country', hue='Cost of Living', data=df)
plt.title('Cost of Living in Different Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Cost of Living')
plt.show()
```
